```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
import undetected_chromedriver as uc
from selenium.webdriver.chrome.options import Options
```

### 整个商品页面

```python
driver = webdriver.Chrome()
```

```python
# 要在中国大陆区访问淘宝
driver.get('https://s.taobao.com/search?catId=100&from=sea_1_searchbutton&page=1&q=cpr%E5%81%87%E4%BA%BA&spm=a2141.241046-cn.searchbar.d_2_searchbox&tmhkh5=&type=p')
time.sleep(5)
```

```python
# 单页所有商品
items = driver.find_element(By.XPATH, '//div[@class="items"]').find_elements(By.XPATH, '//div[@data-category="auctions"]')
```

```python
# 初始化
index = 0
temp_index = 0
page = 0
df = pd.DataFrame(columns=['title', 'location', 'img', 'link'])

# 检查页书
total_page = int(driver.find_element(By.XPATH, '//div[@class="inner clearfix"]//div[@class="total"]').text.split(' ')[1])

while page < 8:
    # 商品信息
    for item in items:
        # 商品链接
        link = item.find_element(By.TAG_NAME, 'a').get_attribute('href')
        # 商品照片
        img = item.find_element(By.TAG_NAME, 'img').get_attribute('src')
        # 商品名称
        title = item.find_element(By.TAG_NAME, 'img').get_attribute('alt')
        df.loc[index, 'title'] = title
        df.loc[index, 'img'] = img
        df.loc[index, 'link'] = link

        index += 1

    # 位置
    locations = driver.find_elements(By.XPATH, '//div[@class="location"]')
    for location in locations:
        df.loc[temp_index, 'location'] = location.text
        temp_index += 1

    # 翻页
    try:
        next_page = driver.find_element(By.XPATH, '//li[@class="item next"]//a[@class="J_Ajax num icon-tag"]')
        next_page.click()
        time.sleep(5)
        items = driver.find_elements(By.XPATH, '//div[@data-category="auctions"]')
    except Exception as e:
        print(f"发生异常：{e}")
        break

    page += 1
```

```python
df.to_csv('taobao.csv', index=False, encoding='utf-8-sig')
df
```
